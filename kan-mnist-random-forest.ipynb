# Data split and Normalization

```python
y = train_df['label']
y.head()
```

```python
x = train_df.iloc[:,1:]
x= x.to_numpy()
print (type(x))
#print (x[0].shape)
print ('before normalize max pixel value', x[0].max())
print ('before normalize  min pixel value', x[0].min())
x= x/255.0
print ('after normalize max pixel value', x[0].max())
print ('after normalize  min pixel value', x[0].min())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)

print ('x_train shape',x_train.shape)
print ('x_test shape',x_test.shape)
print ('y_train shape',y_train.shape)
print ('y_test shape',y_test.shape)
```

# Create Classifier model


```python
from sklearn.ensemble import RandomForestClassifier

random_frst_clf = RandomForestClassifier()
random_frst_clf
```

```python
random_frst_clf.fit(x_train,y_train)
```

```python
y_pred = random_frst_clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print (classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print (accuracy_score(y_test,y_pred))
```
